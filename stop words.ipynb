```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
text = "Cem Filiz, or Cem Adrian with his stage name, was born in Edirne as the second child of a Bosnian family, whose father was a tradesman and mother was a housewife. His real name is Cem Filiz. He stated that he took the surname Adrian, which he used in his professional life, from Hadrianoupolis, the ancient name of the city of Edirne where he was born and raised."
```

```python
stopwords = stopwords.words('english')
```

```python
words = word_tokenize(text)
```

```python
filtered_words = []
for word in words:
  if word not in stopwords:
    filtered_words.append(word)
```
