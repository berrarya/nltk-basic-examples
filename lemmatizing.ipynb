```python
import nltk
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
```

```python
lem = WordNetLemmatizer()
```

```python
words = ['drive', 'driving', 'driver', 'drives', 'drove', 'cats', 'children']
```

```python
for w in words:
  print(lem.lemmatize(w))
```

```python
lem.lemmatize('cats', 'v')
```
